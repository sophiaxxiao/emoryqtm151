(e) Plot by category

<font size = "3">

- Read "data/wdi_urban.csv"<br>
This contains data from 1980 and 2020

- Create a list with the unique values of "year"
- Create a plot with a ...
    - Histogram of "prop_urbanpopulation" <br>
    for each year
    - Set the histogram option ``` alpha = 0.5 ``` for transparency
    - All the histograms should be overlapping
    - Create a legend with the list of years
    - Label the axes and the title

What do we learn from the graph?

Compared to in 1980, in 2020 more countries had a higher proportion of their populations living in cities. This makes sense because development is accompanied by urbanization. 

HINT: See the example at the end of Lecture 10


````python

WDIURBAN = pd.read_csv("data/wdi_urban.csv")

# Get unique categories for "year"
unique_years = WDIURBAN.year.unique()

# Use a for loop to plot a scatter plot between "weight" and "acceleration"
# for each category. Each plot  will have a different color

for category in unique_years:
    df   = WDIURBAN.query("year == @category")
    plt.hist(df["prop_urbanpopulation"], alpha = 0.5)

# Add labels and a legends    
plt.xlabel("Proportion Of Urban Population")
plt.ylabel("Count")
plt.legend(labels = unique_years,
           title  = "Years")
plt.title("Proportion Urban Population Distribution for 1980 and 2020 (from WDI)")

plt.show()
````
