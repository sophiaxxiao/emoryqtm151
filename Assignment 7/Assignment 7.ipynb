<img src="figures/grading_scale.png" alt="drawing" width="600"/>

HINT: See Lecture 13

```python
students = pd.read_csv("data_raw/students.csv")
grade_cutoffs = [0,54,59, 64, 69, 74, 79, 82, 86, 92, 100]
grade_labels = ["F", "D", "C-", "C", "C+", "B-", "B", "B+", "A-", "A"]

students["lettergrade"] = pd.cut(students["numericgrade"],
                                bins = grade_cutoffs,
                                right = True, #include right endpoints
                                labels = grade_labels)

```

- Import the dataset "data_raw/results.csv" to a new object "results"

```python
# Write your own code here


results

objsubset = results.query("milliseconds.str.isnumeric()==False") # create subset of observations with non numeric values for milliseconds variable
nullentry = pd.unique(objsubset["milliseconds"])

results["milliseconds"].replace(nullentry, np.nan, inplace=True) 
# by specifying the inplace argument you modify the original dataframe, replacing nullentry with np.nan

#checking dtypes 
results.query("milliseconds.str.isnumeric()==False")

results[results["milliseconds"].isna()] # checking to see that the object values in "milliseconds" were replaced by NaN

```

(d) Groupby + Aggregate

<font size = "3">

- Compute the mean and standard deviation of "laps", <br>
grouping by "constructorId". Store the new dataset <br>
as "results_agg"
- Sort the aggregate dataset in decending <br>
order of mean laps using ".sort_values()"

HINT: See Lecture 14 <br>


```python
# Write your own code
print(results)
results_agg = results.groupby(["constructorId"]).agg(mean_laps = ('laps', 'mean'),
                                                     stdev_laps = ('laps', 'std')).sort_values(by='mean_laps', ascending=False)

results_agg
```

(d) Query + Groupby + Aggregate

<font size = "3">

- Use ".query()" to subset observations with <br>
'race_id >= 200'
- Compute the mean and standard deviation of "laps", <br>
grouping by "constructorId". Store the new dataset <br>
as "results_agg"

Hint: See Lecture 11 (to refresh ".query()") and <br>
Lecture 14 for how to combine it with ".groupby()"

```python
# Write your own code

subset= results.query('raceId >= 200')

results_agg = subset.groupby(["constructorId"]).agg(mean_laps = ('laps', 'mean'),
                         stdev_laps = ('laps', 'std'))

results_agg
```
